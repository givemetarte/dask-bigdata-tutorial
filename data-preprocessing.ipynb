```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("final-data/csv/preprocessed-flood-damaged-building.csv", encoding="utf-8", dtype=str)
df = df.replace({np.nan : None})
df = df[["관리건축물대장PK","PNU","시군구코드","법정동코드","건물명","새주소도로코드","새주소지상지하코드","동명칭","주택가격","대지면적(㎡)","건축면적(㎡)","지상층수","지하층수","사용승인일","내진설계적용여부"]]
df["URI"] = df["관리건축물대장PK"].apply(lambda x: "http://data.kisti.re.kr/flood-damaged/building/id/" + x)
df = df.set_index("URI").reset_index()
df.to_csv("final-data/csv/final-flood-damaged-building.csv", encoding="utf-8", index=False)

print(len(df.columns))
df.head()
```

```python
df = pd.read_csv("final-data/csv/preprocessed-flood-damaged-cctv.csv", encoding="utf-8", dtype=str)
df = df.replace({np.nan : None})
df["URI"] = df["번호"].apply(lambda x: "http://data.kisti.re.kr/flood-damaged/cctv/id/" + x)
df = df.drop(["읍면동"], axis=1)
df = df.set_index("URI").reset_index()
df.to_csv("final-data/csv/final-flood-damaged-cctv.csv", encoding="utf-8", index=False)

print(df.columns)
print(len(df.columns))
df.head()
```
